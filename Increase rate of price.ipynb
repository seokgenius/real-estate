## 전용면적 구간 나누기

- 65 미만 : 0
- 65 이상 99 미만 : 1
- 99 이상 135 미만 : 2
- 135 이상 165 미만 : 3
- 165 이상 : 4

```python
# 2016년도
apart16_df.loc[apart16_df['area'] < 65, 'area'] = 0
apart16_df.loc[(apart16_df['area'] >= 65) & (apart16_df['area'] < 99) , 'area'] = 1
apart16_df.loc[(apart16_df['area'] >= 99) & (apart16_df['area'] < 135) , 'area'] = 2
apart16_df.loc[(apart16_df['area'] >= 135) & (apart16_df['area'] < 165) , 'area'] = 3
apart16_df.loc[apart16_df['area'] >= 165, 'area'] = 4

# 2017년도
apart17_df.loc[apart17_df['area'] < 65, 'area'] = 0
apart17_df.loc[(apart17_df['area'] >= 65) & (apart17_df['area'] < 99) , 'area'] = 1
apart17_df.loc[(apart17_df['area'] >= 99) & (apart17_df['area'] < 135) , 'area'] = 2
apart17_df.loc[(apart17_df['area'] >= 135) & (apart17_df['area'] < 165) , 'area'] = 3
apart17_df.loc[apart17_df['area'] >= 165, 'area'] = 4

# 2018년도
apart18_df.loc[apart18_df['area'] < 65, 'area'] = 0
apart18_df.loc[(apart18_df['area'] >= 65) & (apart18_df['area'] < 99) , 'area'] = 1
apart18_df.loc[(apart18_df['area'] >= 99) & (apart18_df['area'] < 135) , 'area'] = 2
apart18_df.loc[(apart18_df['area'] >= 135) & (apart18_df['area'] < 165) , 'area'] = 3
apart18_df.loc[apart18_df['area'] >= 165, 'area'] = 4

# 2019년도
apart19_df.loc[apart19_df['area'] < 65, 'area'] = 0
apart19_df.loc[(apart19_df['area'] >= 65) & (apart19_df['area'] < 99) , 'area'] = 1
apart19_df.loc[(apart19_df['area'] >= 99) & (apart19_df['area'] < 135) , 'area'] = 2
apart19_df.loc[(apart19_df['area'] >= 135) & (apart19_df['area'] < 165) , 'area'] = 3
apart19_df.loc[apart19_df['area'] >= 165, 'area'] = 4


apart18_df
```
