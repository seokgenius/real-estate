# 2016-19년도 면적별 거래금액 분석

서울시 25개구의 면적별 거래금액 평균을 년도별로 시각화

전용면적 구간
- 65 이하 : 0
- 65 이상 99 미만 : 1
- 99 이상 135 미만 : 2
- 135 이상 165 미만 : 3
- 165 이상 : 4

```python
import pandas as pd
import numpy as np
```

```python
apart16_df = pd.read_csv('apart16.csv', encoding='CP949')
apart17_df = pd.read_csv('apart17.csv', encoding='CP949')
apart18_df = pd.read_csv('apart18.csv', encoding='CP949')
apart19_df = pd.read_csv('apart19.csv', encoding='CP949')
```

### 자치구 추출

```python
apart16_df['address'] = [gu.split()[1] for gu in apart16_df['address']]
apart17_df['address'] = [gu.split()[1] for gu in apart17_df['address']]
apart18_df['address'] = [gu.split()[1] for gu in apart18_df['address']]
apart19_df['address'] = [gu.split()[1] for gu in apart19_df['address']]

apart16_df
```

### 년도 삽입

```python
apart16_df['year'] = 2016
apart16_df = apart16_df[['year', 'address', 'area', 'amount']]

apart17_df['year'] = 2017
apart17_df = apart17_df[['year', 'address', 'area', 'amount']]

apart18_df['year'] = 2018
apart18_df = apart18_df[['year', 'address', 'area', 'amount']]

apart19_df['year'] = 2019
apart19_df = apart19_df[['year', 'address', 'area', 'amount']]

apart17_df
```

### 전용면적 나누기

```python
# 2016년도
apart16_df.loc[apart16_df['area'] < 65, 'area'] = 0
apart16_df.loc[(apart16_df['area'] >= 65) & (apart16_df['area'] < 99) , 'area'] = 1
apart16_df.loc[(apart16_df['area'] >= 99) & (apart16_df['area'] < 135) , 'area'] = 2
apart16_df.loc[(apart16_df['area'] >= 135) & (apart16_df['area'] < 165) , 'area'] = 3
apart16_df.loc[apart16_df['area'] >= 165, 'area'] = 4

# 2017년도
apart17_df.loc[apart17_df['area'] < 65, 'area'] = 0
apart17_df.loc[(apart17_df['area'] >= 65) & (apart17_df['area'] < 99) , 'area'] = 1
apart17_df.loc[(apart17_df['area'] >= 99) & (apart17_df['area'] < 135) , 'area'] = 2
apart17_df.loc[(apart17_df['area'] >= 135) & (apart17_df['area'] < 165) , 'area'] = 3
apart17_df.loc[apart17_df['area'] >= 165, 'area'] = 4

# 2018년도
apart18_df.loc[apart18_df['area'] < 65, 'area'] = 0
apart18_df.loc[(apart18_df['area'] >= 65) & (apart18_df['area'] < 99) , 'area'] = 1
apart18_df.loc[(apart18_df['area'] >= 99) & (apart18_df['area'] < 135) , 'area'] = 2
apart18_df.loc[(apart18_df['area'] >= 135) & (apart18_df['area'] < 165) , 'area'] = 3
apart18_df.loc[apart18_df['area'] >= 165, 'area'] = 4

# 2019년도
apart19_df.loc[apart19_df['area'] < 65, 'area'] = 0
apart19_df.loc[(apart19_df['area'] >= 65) & (apart19_df['area'] < 99) , 'area'] = 1
apart19_df.loc[(apart19_df['area'] >= 99) & (apart19_df['area'] < 135) , 'area'] = 2
apart19_df.loc[(apart19_df['area'] >= 135) & (apart19_df['area'] < 165) , 'area'] = 3
apart19_df.loc[apart19_df['area'] >= 165, 'area'] = 4

apart18_df
```

### 2016-19년 데이터 하나로 묶기

```python
entire_df = pd.concat([apart16_df,apart17_df,apart18_df,apart19_df])
entire_df
```

### 년도별, 자치구별, 전용면적별 평균 거래금액구하기

```python
mean_amount = entire_df.pivot_table('amount', index=['year', 'address'], columns='area', aggfunc='mean')
mean_amount = mean_amount.fillna(0)
mean_amount
```

### 2016년도와 2019년 기준으로 변동율 구하기

```python
mean16 = apart16_df.pivot_table('amount', index=['year', 'address'], columns='area', aggfunc='mean')
mean16 = mean16.fillna(1)
mean16
```

```python
mean19 = apart19_df.pivot_table('amount', index=['year', 'address'], columns='area', aggfunc='mean')
mean19 = mean19.fillna(1)
mean19
```

```python
# 전용면적 구간 1
increaseRate1 = ((mean19.loc[2019, '강남구'][1]-mean16.loc[2016, '강남구'][1])/mean16.loc[2016, '강남구'][1])*100
increaseRate1
```

```python
# 전용면적 구간 2
increaseRate2 = ((mean19.loc[2019, '강남구'][2]-mean16.loc[2016, '강남구'][2])/mean16.loc[2016, '강남구'][2])*100
increaseRate2
```

```python
# 전용면적 구간 3
increaseRate3 = ((mean19.loc[2019, '강남구'][3]-mean16.loc[2016, '강남구'][3])/mean16.loc[2016, '강남구'][3])*100
increaseRate3
```

```python
# 전용면적 구간 4
increaseRate4 = ((mean19.loc[2019, '강남구'][4]-mean16.loc[2016, '강남구'][4])/mean16.loc[2016, '강남구'][4])*100
increaseRate4
```

```python
# 강남구 거래금액 상승률 평균
increaseRate_mean = (increaseRate0+increaseRate1+increaseRate2+increaseRate3+increaseRate4)/5
increaseRate_mean
```
