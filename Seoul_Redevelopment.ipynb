### 1. 데이터 정리하기

```python
import pandas as pd
```

```python
seoul_re = pd.read_csv('Seoul_Re.csv')
seoul_re
```

#### 1-1 동을 기준으로 데이터를 쪼갠 후 '00동'만 리스트에 넣기 (동별 시각화에 사용, 선택사항)

```python
seoul_re['대표지번']
```

```python
address = []
words = seoul_re['대표지번']

for i in words:
    words = seoul_re['대표지번'].str.split( )
    i = i.split( )
    address.append(i[0])
print(address)
```

```python
seoul_re['대표지번'] = address
seoul_re
```

#### 1-2 열이름 변경 및 필요없는 데이터 삭제하기

```python
# 사업번호(number) 삭제
seoul_re.drop(["사업번호"], axis='columns', inplace=True)
seoul_re.drop(["상태"], axis='columns', inplace=True)

# 열이름 변경
seoul_re.rename(columns={'자치구': 'district', '사업구분':'category', '대표지번':'address'}, inplace=True)
seoul_re
```

### 4. 자치구별 서울시 지도 시각화

#### 4-1 자치구별 '도시환경정비' 갯수

```python
import folium

geo_json = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

raw_data1 = [
    ['강동구',2],
    ['강북구',3],
    ['광진구',1],
    ['동대문구',3],
    ['동작구',1],
    ['마포구',2],
    ['서대문구',4],
    ['성동구',1],
    ['성북구', 2],
    ['양천구',2],
    ['영등포구', 34],
    ['용산구',10],
    ['은평구',2],
    ['종로구',6],
    ['중구',1],
    ['중랑구',2],
    ['관악구', 9],
    ['구로구', 1],
    ['금천구', 1],
    ['노원구', 6],
    ['도봉구', 2],
    ['송파구', 5],
    ['강남구', 0],
    ['강서구',0],
    ['서초구',0]]

data = pd.DataFrame(raw_data1, columns=['name','value'])
data


m = folium.Map(location=[37.566345, 126.977893])

folium.Choropleth(
    geo_data = geo_json,
    name = 'choropleth',
    data=data,
    columns=['name','value'],
    key_on = 'feature.properties.name',
    fill_color = 'PuBu',
    fill_opacity = 0.7, 
    line_opaciity= 0.2,
    legend_name='자치구별 도시환경정비 개수 (개)'

).add_to(m)

m

```
