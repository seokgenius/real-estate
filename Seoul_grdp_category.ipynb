# 자치구별 지역총생산을 네가지 카테고리로 분류하기 

```python
import pandas as pd
import numpy as np


grdp_df = pd.read_csv('grdp_df.csv', thousands=',')
grdp_df
```

### GRDP 구분하기

```python
# 5000000이하  GRDP 0 
# 5000000이상 10000000이하 GRDP 1 
# 10000000이상 20000000이하 GRDP 2 
# 20000000이상 30000000이하 GRDP 3 
# 30000000이상 GRDP 4

# 평균값 14377596
# 중앙값 7452206
# 최솟값 2816851
# 최댓값 59981514

grdp_df.loc[grdp_df['GRDP'] < 5000000.0, 'GRDP'] = 0
grdp_df.loc[((grdp_df['GRDP'] >=5000000.0 ) & (grdp_df['GRDP'] < 10000000.0)) , 'GRDP'] = 1
grdp_df.loc[((grdp_df['GRDP'] >=10000000.0) & (grdp_df['GRDP'] < 20000000.0)) , 'GRDP'] = 2
grdp_df.loc[((grdp_df['GRDP'] >=20000000.0) & (grdp_df['GRDP'] < 30000000.0)) , 'GRDP'] = 3
grdp_df.loc[grdp_df['GRDP'] >= 30000000.0 , 'GRDP'] = 4

grdp_df
```

```python
# DataFrame으로 저장하기
grdp_df.to_csv('grdp_cate.csv', encoding='utf-8-sig', index=False)
```
