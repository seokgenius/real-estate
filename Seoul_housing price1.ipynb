### 년도별 데이터 불러오기

```python
import pandas as pd
import numpy as np

apart16 = pd.read_csv('apart16_sell.csv')
apart17 = pd.read_csv('apart17_sell.csv')
apart18 = pd.read_csv('apart18_sell.csv')
apart19 = pd.read_csv('apart19_sell.csv')
```

### 1. 2016년도 데이터 정리

```python
# DataFrame으로 변경
apart16_df = pd.DataFrame(apart16)
apart16_df.head()
```

#### 1-1. 열이름 변경 및 필요없는 데이터 삭제하기 

```python
# 번지, 본번, 부번, 단지명, 계약년월, 계약일, 층, 건축년도, 도로명 삭제
apart16_df.drop(["번지"], axis='columns', inplace=True)
apart16_df.drop(["본번"], axis='columns', inplace=True)
apart16_df.drop(["부번"], axis='columns', inplace=True)
apart16_df.drop(["단지명"], axis='columns', inplace=True)
apart16_df.drop(["계약년월"], axis='columns', inplace=True)
apart16_df.drop(["계약일"], axis='columns', inplace=True)
apart16_df.drop(["층"], axis='columns', inplace=True)
apart16_df.drop(["건축년도"], axis='columns', inplace=True)
apart16_df.drop(["도로명"], axis='columns', inplace=True)
```

```python
apart16_df.head()
```

#### 1-2. '시군구'를 '자치구'와  '동'으로 나누기

```python
gudong16_list = pd.DataFrame({
    '자치구' : apart16_df['시군구'],
    '동' : apart16_df['시군구'],
    '전용면적' : apart16_df['전용면적(㎡)'],
    '거래금액(만원)' : apart16_df['거래금액(만원)']
})

gudong16_list['자치구'] = [each_gu.split()[1] for each_gu in apart16_df['시군구']]
gudong16_list['동'] = [each_dong.split()[2] for each_dong in apart16_df['시군구']]

# object -> list 
gudong16_list['거래금액(만원)'] = gudong16_list['거래금액(만원)'].replace('[\,]', '', regex=True)
gudong16_list['거래금액(만원)'] = pd.to_numeric(gudong16_list['거래금액(만원)'])

gudong16_list
```

```python
# 데이터 프레임으로 저장
gudong16_list.to_csv('apart16_df.csv', encoding='utf-8-sig', index=False)
```
