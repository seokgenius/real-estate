### 년도별 데이터 불러오기

```python
import pandas as pd
import numpy as np

apart16_df = pd.read_csv('apart16_df.csv')
apart17_df = pd.read_csv('apart17_df.csv')
apart18_df = pd.read_csv('apart18_df.csv')
apart19_df = pd.read_csv('apart19_df.csv')
```

### 1. 2016년 ~ 2019년 데이터 정리

```python
apart16_df
```

```python
# 16년도 데이터 정리
apart16_df1 = pd.DataFrame({
                '기간' : 2016,
                '구' : apart16_df['자치구'],
                '전용면적' : apart16_df['전용면적'],
                '거래금액' : apart16_df['거래금액(만원)']
})

# 17년도 데이터 정리
apart17_df1 = pd.DataFrame({
                '기간' : 2017,
                '구' : apart17_df['자치구'],
                '전용면적' : apart17_df['전용면적'],
                '거래금액' : apart16_df['거래금액(만원)']
})

# 16년도 데이터 정리
apart18_df1 = pd.DataFrame({
                '기간' : 2018,
                '구' : apart18_df['자치구'],
                '전용면적' : apart18_df['전용면적'],
                '거래금액' : apart18_df['거래금액(만원)']
})


# 19년도 데이터 정리
apart19_df1 = pd.DataFrame({
                '기간' : 2019,
                '구' : apart19_df['자치구'],
                '전용면적' : apart19_df['전용면적'],
                '거래금액' : apart19_df['거래금액(만원)']
})

apart16_df1.head(20)
```

### 2. 기준 년도별 전용 면적 구분하기

```python
# 65이하 전용면적 0 
# 65이상 99미만  전용면적 1
# 99이상 135미만 전용면적 2
# 135이상 165미만 전용면적 3
# 165이상 전용면적 4


# 16년도
apart16_df1.loc[apart16_df1['전용면적'] < 65.0, '전용면적'] = 0
apart16_df1.loc[((apart16_df1['전용면적'] >=65.0 ) & (apart16_df1['전용면적'] < 99.0)) , '전용면적'] = 1
apart16_df1.loc[((apart16_df1['전용면적'] >=99.0) & (apart16_df1['전용면적'] < 135.0)) , '전용면적'] = 2
apart16_df1.loc[((apart16_df1['전용면적'] >=135.0) & (apart16_df1['전용면적'] < 165.0)) , '전용면적'] = 3
apart16_df1.loc[apart16_df1['전용면적'] >= 165.0 , '전용면적'] = 4

# 17년도
apart17_df1.loc[apart17_df1['전용면적'] < 65.0, '전용면적'] = 0
apart17_df1.loc[((apart17_df1['전용면적'] >=65.0 ) & (apart17_df1['전용면적'] < 99.0)) , '전용면적'] = 1
apart17_df1.loc[((apart17_df1['전용면적'] >=99.0) & (apart17_df1['전용면적'] < 135.0)) , '전용면적'] = 2
apart17_df1.loc[((apart17_df1['전용면적'] >=135.0) & (apart17_df1['전용면적'] < 165.0)) , '전용면적'] = 3
apart17_df1.loc[apart17_df1['전용면적'] >= 165.0 , '전용면적'] = 4

# 16년도
apart18_df1.loc[apart18_df1['전용면적'] < 65.0, '전용면적'] = 0
apart18_df1.loc[((apart18_df1['전용면적'] >=65.0 ) & (apart18_df1['전용면적'] < 99.0)) , '전용면적'] = 1
apart18_df1.loc[((apart18_df1['전용면적'] >=99.0) & (apart18_df1['전용면적'] < 135.0)) , '전용면적'] = 2
apart18_df1.loc[((apart18_df1['전용면적'] >=135.0) & (apart18_df1['전용면적'] < 165.0)) , '전용면적'] = 3
apart18_df1.loc[apart18_df1['전용면적'] >= 165.0 , '전용면적'] = 4

# 19년도
apart19_df1.loc[apart19_df1['전용면적'] < 65.0, '전용면적'] = 0
apart19_df1.loc[((apart19_df1['전용면적'] >=65.0 ) & (apart19_df1['전용면적'] < 99.0)) , '전용면적'] = 1
apart19_df1.loc[((apart19_df1['전용면적'] >=99.0) & (apart19_df1['전용면적'] < 135.0)) , '전용면적'] = 2
apart19_df1.loc[((apart19_df1['전용면적'] >=135.0) & (apart19_df1['전용면적'] < 165.0)) , '전용면적'] = 3
apart19_df1.loc[apart19_df1['전용면적'] >= 165.0 , '전용면적'] = 4


apart16_df1.head(20)

```

```python
tot_df = pd.concat([apart16_df1,apart17_df1,apart18_df1,apart19_df1])
tot_df
```

### 3. 기간별, 구별, 전용면적별 거래금액 그룹화

```python
# 카테고리별 그룹화 = 

cate1_df = tot_df[ tot_df['구'] == '강남구']
cate1_df.head(30)

cate1_df_agg1 = cate1_df.groupby(['기간','구','전용면적']).agg({
    '거래금액':'mean'
})

cate1_df_agg1
tot_df2 = cate1_df_agg1.transpose()
tot_df2
```

```python
# 카테고리별 그룹화 = 서초구
cate2_df =  apart16_df1[ apart16_df1['구'] == '서초구']

apart16_df_agg2 = cate2_df.groupby(['구','전용면적']).agg({
    '거래금액':'mean'
})

apart16_df3 = apart16_df_agg2.transpose()
apart16_df3
```

```python
# 2016년 구별 그룹화
apart16_df_agg = apart16_df1.groupby(['기간','구','전용면적']).agg({
    '거래금액':'mean'
})


# 2017년 구별 그룹화
apart17_df_agg = apart17_df1.groupby(['기간','구','전용면적']).agg({
    '거래금액':'mean'
})

# 2018년 구별 그룹화
apart18_df_agg = apart18_df1.groupby(['기간','구','전용면적']).agg({
    '거래금액':'mean'
})

# 2019년 구별 그룹화
apart19_df_agg = apart19_df1.groupby(['기간','구','전용면적']).agg({
    '거래금액':'mean'
})

apart16_df_agg
```
