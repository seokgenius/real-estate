```python
import pandas as pd
import numpy as np
```

```python
aca_df = pd.read_csv('academy_2018.csv', encoding='euckr', thousands=',')
aca_df
```

```python
aca_df.info()
```

```python
aca_2018 = aca_df.sort_values(by=['계'], ascending=False)
aca_2018
```

```python
# 200미만 학원갯수 0 
# 200이상 300미만 학원갯수 1
# 300이상 500미만 학원갯수 2
# 500이상 1000미만 학원갯수 3
# 1000이상 학원갯수 4

aca_2018.loc[aca_2018['계'] < 200, '계'] = 0
aca_2018.loc[((aca_2018['계'] >=200 ) & (aca_2018['계'] < 300)) , '계'] = 1
aca_2018.loc[((aca_2018['계'] >=300) & (aca_2018['계'] < 500)) , '계'] = 2
aca_2018.loc[((aca_2018['계'] >=500) & (aca_2018['계'] < 1000)) , '계'] = 3
aca_2018.loc[aca_2018['계'] >= 1000 , '계'] = 4

aca_2018
```

```python
aca_2018.rename(columns = {'계' : '학원수'}, inplace = True)
```

```python
aca_2018.to_csv("aca_2018.csv",index=False, encoding='utf-8-sig')
```
