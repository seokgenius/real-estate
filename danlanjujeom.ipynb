```python
import pandas as pd
```

```python
bar_df = pd.read_csv('bar.csv', encoding='CP949')
bar_df.head()
```

```python
gu_list = []

for i in range(0, len(bar_df)):
    if type(bar_df['oldAddress'][i]) == float:
        addr = bar_df['newAddress'][i].split()
        gu_list.append(addr[1])
    else:
        addr = bar_df['oldAddress'][i].split()
        gu_list.append(addr[1])
        
print(gu_list)
```

```python
bar_df['bar'] = pd.Series(gu_list)
```

```python
bar_df
```

```python
bar_df['bar'].value_counts()
```

```python
bar_number = pd.DataFrame(bar_df['bar'].value_counts())
bar_number['region'] = bar_number.index
bar_number = bar_number[['region', 'bar']]
bar_number
```

```python
bar_number.to_csv('danlanjujeom_raw.csv', index=False, encoding='utf-8-sig')
```

스케일 조정

- ~ 65 : 0
- 66 ~ 95 : 0.8
- 96 ~ 110 : 1.6
- 111 ~ 135 : 2.4
- 135 ~ 170 : 3.2
- 171 ~ : 4.0

```python
bar_number.loc[bar_number['bar'] <= 65, 'bar'] = 0
bar_number.loc[(bar_number['bar'] > 65) & (bar_number['bar'] <= 95) , 'bar'] = 0.8
bar_number.loc[(bar_number['bar'] > 95) & (bar_number['bar'] <= 110) , 'bar'] = 1.6
bar_number.loc[(bar_number['bar'] > 110) & (bar_number['bar'] <= 135) , 'bar'] = 2.4
bar_number.loc[(bar_number['bar'] > 135) & (bar_number['bar'] <= 170) , 'bar'] = 3.2
bar_number.loc[bar_number['bar'] > 170, 'bar'] = 4
```
