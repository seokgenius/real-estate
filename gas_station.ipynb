```python
import pandas as pd
```

```python
gas_df = pd.read_csv('gas.csv', encoding='CP949')
gas_df.head()
```

```python
gu_list = []
addr = gas_df['region']

for i in addr:
    addr = gas_df['region'].str.split( )
    i = i.split( )
    gu_list.append(i[1])
print(gu_list)
```

```python
gas_df['gas'] = pd.Series(gu_list)
```

```python
gas_df
```

```python
gas_df['gas'].value_counts()
```

```python
gas_number = pd.DataFrame(gas_df['gas'].value_counts())
gas_number['region'] = gas_number.index
gas_number = gas_number[['region','gas']]
gas_number
```

```python
gas_number.to_csv('gas_raw.csv', index=False, encoding='utf-8-sig')
```

스케일 조정

- ~2 : 0
- 3 : 1
- 4 : 2
- 5~6 : 3
- 6~ : 4

```python
gas_number.loc[gas_number['gas'] <= 2, 'gas'] = 0
gas_number.loc[(gas_number['gas'] > 2) & (gas_number['gas'] <= 3), 'gas'] = 1
gas_number.loc[(gas_number['gas'] > 3) & (gas_number['gas'] <= 4), 'gas'] = 2
gas_number.loc[(gas_number['gas'] > 4) & (gas_number['gas'] <= 6) , 'gas'] = 3
gas_number.loc[gas_number['gas'] > 6, 'gas'] = 4
```

```python
gas_number['gas'].value_counts()
```
