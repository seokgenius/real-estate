```python
import pandas as pd
```

```python
hospital_df = pd.read_csv('hospital.csv', encoding='CP949')
hospital_df.head()
```

```python
gu_list = []

for i in range(0, len(hospital_df)):
    if type(hospital_df['oldAddress'][i]) == float:
        addr = hospital_df['newAddress'][i].split()
        gu_list.append(addr[1])
    else:
        addr = hospital_df['oldAddress'][i].split()
        gu_list.append(addr[1])
        
print(gu_list)
```

```python
hospital_df['hospital'] = pd.Series(gu_list)
```

```python
hospital_df
```

```python
hospital_df['hospital'].value_counts()
```

```python
hospital_number = pd.DataFrame(hospital_df['hospital'].value_counts())
hospital_number['region'] = hospital_number.index
hospital_number = hospital_number[['region', 'hospital']]
hospital_number
```

```python
hospital_number.to_csv('hospital_raw.csv', index=False, encoding='utf-8-sig')
```

스케일 조정

- ~ 9 : 0
- 10 ~ 13 : 1
- 14 ~ 20 : 2
- 21 ~ 31 : 3
- 32 ~ : 4

```python
hospital_number.loc[hospital_number['hospital'] <= 9, 'hospital'] = 0
hospital_number.loc[(hospital_number['hospital'] > 9) & (hospital_number['hospital'] <= 13) , 'hospital'] = 1
hospital_number.loc[(hospital_number['hospital'] > 13) & (hospital_number['hospital'] <= 20) , 'hospital'] = 2
hospital_number.loc[(hospital_number['hospital'] > 20) & (hospital_number['hospital'] <= 31) , 'hospital'] = 3
hospital_number.loc[hospital_number['hospital'] > 31, 'hospital'] = 4
```

```python
hospital_number['hospital'].value_counts()
```
