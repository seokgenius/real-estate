```python
import pandas as pd
```

```python
park_df = pd.read_csv('park_first.csv', encoding='CP949')
park_df.head()
```

```python
park_df['region'].value_counts()
```

```python
park_number = pd.DataFrame(park_df['region'].value_counts())
park_number['gu'] = park_number.index
park_number = park_number[['gu','region']]
park_number.columns=['region', 'park']
park_number
```

```python
park_number.to_csv('park_raw.csv', index=False, encoding='utf-8-sig')
```

스케일 조정

- ~ 3 : 0
- 4 : 1
- 5 ~ 6 : 2
- 7 ~ 8 : 3
- 9 ~ : 4

```python
park_number.loc[park_number['park'] <= 3, 'park'] = 0
park_number.loc[(park_number['park'] > 3) & (park_number['park'] <= 4) , 'park'] = 1
park_number.loc[(park_number['park'] > 4) & (park_number['park'] <= 6) , 'park'] = 2
park_number.loc[(park_number['park'] > 6) & (park_number['park'] <= 8) , 'park'] = 3
park_number.loc[park_number['park'] > 8, 'park'] = 4
```
