```python
import pickle
from konlpy.tag import *

okt = Okt()

def tokenizer(text):
    return okt.morphs(text)
```

```python
with open('population_article_pipe.dat', 'rb') as fp :
    pipe = pickle.load(fp)
```

```python
text = input('기사를 작성해주세요 : ')

str1 = [text]

result1 = pipe.predict(str1)
result2 = pipe.predict_proba(str1)
print(result2)

if result1[0] == '1' :
    print('부동산 가격 상승 예정')
else :
    print('부동산 가격 하락 예정')
```
