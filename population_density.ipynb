```python
import pandas as pd
import numpy as np
```

```python
pop_df = pd.read_csv('seoul_pop.csv', encoding='euckr', thousands=',')
pop_df.head()
```

```python
pop_2018 = pop_df[(pop_df['기간']==2018) & (pop_df['동']=='소계')].sort_values(by=['인구밀도'], ascending=False)
pop_2018
```

```python
pop_2018 = pop_2018.drop("기간", axis=1)
pop_2018 = pop_2018.drop("동", axis=1)
pop_2018 = pop_2018.drop("인구", axis=1)
pop_2018 = pop_2018.drop("면적", axis=1)
```

```python
pop_2018.info()
```

```python
# 10000미만 인구밀도 0 
# 10000이상 15000미만 인구밀도 1
# 15000이상 20000미만 인구밀도 2
# 20000이상 25000미만 인구밀도 3
# 25000이상 인구밀도 4

pop_2018.loc[pop_2018['인구밀도'] < 10000, '인구밀도'] = 0
pop_2018.loc[((pop_2018['인구밀도'] >=10000 ) & (pop_2018['인구밀도'] < 15000)) , '인구밀도'] = 1
pop_2018.loc[((pop_2018['인구밀도'] >=15000) & (pop_2018['인구밀도'] < 20000)) , '인구밀도'] = 2
pop_2018.loc[((pop_2018['인구밀도'] >=20000) & (pop_2018['인구밀도'] < 25000)) , '인구밀도'] = 3
pop_2018.loc[pop_2018['인구밀도'] >= 25000 , '인구밀도'] = 4
```
