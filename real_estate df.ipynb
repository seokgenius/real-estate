```python
import pandas as pd
```

```python
edu_df = pd.read_csv('aca_2018_test.csv')
edu_df.index = edu_df['region']
del edu_df['region']

bar_df = pd.read_csv('danlanjujeom_test.csv')
bar_df.index = bar_df['region']
del bar_df['region']

gas_df = pd.read_csv('gas_test.csv')
gas_df.index = gas_df['region']
del gas_df['region']

hospital_df = pd.read_csv('hospital_test.csv')
hospital_df.index = hospital_df['region']
del hospital_df['region']

park_df = pd.read_csv('park_test.csv')
park_df.index = park_df['region']
del park_df['region']

population_df = pd.read_csv('pop_density_2018_test.csv')
population_df.index = population_df['region']
del population_df['region']

store_df = pd.read_csv('store_test.csv')
store_df.index = store_df['region']
del store_df['region']

theater_df = pd.read_csv('theater_test.csv')
theater_df.index = theater_df['region']
del theater_df['region']

saloon_df = pd.read_csv('uheungjujeom_test.csv')
saloon_df.index = saloon_df['region']
del saloon_df['region']

waste_df = pd.read_csv('waste_test.csv')
waste_df.index = waste_df['region']
del waste_df['region']

y_data = pd.read_csv('increase_rate.csv')
y_data.index = y_data['region']
del y_data['region']

edu_df
```

```python
real_estate_df = pd.concat([edu_df, bar_df, gas_df, hospital_df, park_df, population_df, store_df,
        theater_df, saloon_df, waste_df, y_data], axis=1, ignore_index=False)
real_estate_df
```

```python
real_estate_df = real_estate_df.fillna(0)
real_estate_df
```

```python
real_estate_df.to_csv('real_estate_df.csv', index=False, encoding='utf-8-sig')
```
