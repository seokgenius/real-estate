```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
from matplotlib import rcParams
plt.rc('font', family='Malgun Gothic')
plt.rc('font', size='10')
```

```python
rebuild_df = pd.read_csv('rebuild.csv', encoding='euckr')
rebuild_df.head()
```

```python
rebuild_df.info()
```

```python
rebuild_count_df = pd.DataFrame(rebuild_df['자치구'].value_counts())
rebuild_count_df.rename(columns=lambda x: x.replace('자치구', '재개발'), inplace=True)
rebuild_count_df
```

```python
rebuild_count_df['자치구'] = rebuild_count_df.index
```

```python
rebuild_count_df = rebuild_count_df[['자치구','재개발']]
```

```python
rebuild_count_df
```

```python
rebuild_count_df.to_csv('rebuild_count_raw.csv', index=False, encoding='utf-8-sig')
```

```python
rebuild_count_df.loc[rebuild_count_df['재개발'] < 10, '재개발'] = 0
rebuild_count_df.loc[(rebuild_count_df['재개발'] >= 10) & (rebuild_count_df['재개발'] < 15) , '재개발'] = 1
rebuild_count_df.loc[(rebuild_count_df['재개발'] >= 15) & (rebuild_count_df['재개발'] < 25) , '재개발'] = 2
rebuild_count_df.loc[(rebuild_count_df['재개발'] >= 25) & (rebuild_count_df['재개발'] < 40) , '재개발'] = 3
rebuild_count_df.loc[rebuild_count_df['재개발'] >= 40, '재개발'] = 4
```

```python
rebuild_count_df
```

```python
rebuild_count_df.to_csv('rebuild_count.csv', index=False, encoding='utf-8-sig')
```
