```python
import pandas as pd
```

```python
store_df = pd.read_csv('store.csv', encoding = 'CP949')
store_df.head()
```

```python
gu_list = []

for i in range(0, len(store_df)):
    if type(store_df['oldAddress'][i]) == float:
        addr = store_df['newAddress'][i].split()
        gu_list.append(addr[1])
    else:
        addr = store_df['oldAddress'][i].split()
        gu_list.append(addr[1])

print(gu_list)
```

```python
store_df['store'] = pd.Series(gu_list)
```

```python
store_df
```

```python
store_df['store'].value_counts()
```

```python
store_number = pd.DataFrame(store_df['store'].value_counts())
store_number['region'] = store_number.index
store_number = store_number[['region','store']]
store_number
```

```python
store_number.to_csv('store_raw.csv', index=False, encoding='utf-8-sig')
```

스케일 조정

- ~ 20 : 0
- 21 ~ 25 : 1
- 26 ~ 30 : 2
- 31 ~ 45 : 3
- 46 ~ : 4

```python
store_number.loc[store_number['store'] <= 20, 'store'] = 0
store_number.loc[(store_number['store'] > 20) & (store_number['store'] <= 25) , 'store'] = 1
store_number.loc[(store_number['store'] > 25) & (store_number['store'] <= 30) , 'store'] = 2
store_number.loc[(store_number['store'] > 30) & (store_number['store'] <= 45) , 'store'] = 3
store_number.loc[store_number['store'] > 45, 'store'] = 4
```
