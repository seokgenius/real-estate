```python
import pandas as pd
```

```python
theater_df = pd.read_csv('theater.csv', encoding='CP949')
theater_df.head()
```

```python
gu_list = []

for i in range(0, len(theater_df)):
    if type(theater_df['oldAddress'][i]) == float:
        addr = theater_df['newAddress'][i].split()
        gu_list.append(addr[1])
    else:
        addr = theater_df['oldAddress'][i].split()
        gu_list.append(addr[1])

print(gu_list)
```

```python
theater_df['theater'] = pd.Series(gu_list)
```

```python
theater_df
```

```python
theater_df['theater'].value_counts()
```

```python
theater_number = pd.DataFrame(theater_df['theater'].value_counts())
theater_number['region'] = theater_number.index
theater_number = theater_number[['region','theater']]
theater_number
```

```python
theater_number.to_csv('theater_raw.csv', index=False, encoding='utf-8-sig')
```

스케일 조정

- 1 : 0
- 2 : 1
- 3 : 2
- 4~5 : 3
- 5~ : 4

```python
theater_number.loc[theater_number['theater'] <= 1, 'theater'] = 0
theater_number.loc[theater_number['theater'] == 2 , 'theater'] = 1
theater_number.loc[theater_number['theater'] == 3 , 'theater'] = 2
theater_number.loc[(theater_number['theater'] > 3) & (theater_number['theater'] <= 5) , 'theater'] = 3
theater_number.loc[theater_number['theater'] > 5, 'theater'] = 4
```

```python
theater_number['theater'].value_counts()
```
