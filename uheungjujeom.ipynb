```python
import pandas as pd
```

```python
saloon_df = pd.read_csv('saloon.csv', encoding='CP949')
saloon_df.head()
```

```python
gu_list = []

for i in range(0, len(saloon_df)):
    if type(saloon_df['oldAddress'][i]) == float:
        addr = saloon_df['newAddress'][i].split()
        gu_list.append(addr[1])
    else:
        addr = saloon_df['oldAddress'][i].split()
        gu_list.append(addr[1])
        
print(gu_list)
```

```python
saloon_df['saloon'] = pd.Series(gu_list)
```

```python
saloon_df
```

```python
saloon_df['saloon'].value_counts()
```

```python
saloon_number = pd.DataFrame(saloon_df['saloon'].value_counts())
saloon_number['region'] = saloon_number.index
saloon_number = saloon_number[['region', 'saloon']]
saloon_number
```

```python
saloon_number.to_csv('uheungjujeom_raw.csv', index=False, encoding='utf-8-sig')
```

스케일 조정

- ~ 8 : 0
- 9 ~ 35 : 0.6
- 36 ~ 50 : 1.2
- 51 ~ 75 : 1.9
- 76 ~ 95 : 2.6
- 96 ~ 200 : 3.3
- 200 ~  : 4

```python
saloon_number.loc[saloon_number['saloon'] <= 8, 'saloon'] = 0
saloon_number.loc[(saloon_number['saloon'] > 8) & (saloon_number['saloon'] <= 35) , 'saloon'] = 0.6
saloon_number.loc[(saloon_number['saloon'] > 35) & (saloon_number['saloon'] <= 50) , 'saloon'] = 1.2
saloon_number.loc[(saloon_number['saloon'] > 50) & (saloon_number['saloon'] <= 75) , 'saloon'] = 1.9
saloon_number.loc[(saloon_number['saloon'] > 75) & (saloon_number['saloon'] <= 95) , 'saloon'] = 2.6
saloon_number.loc[(saloon_number['saloon'] > 95) & (saloon_number['saloon'] <= 200) , 'saloon'] = 3.3
saloon_number.loc[saloon_number['saloon'] > 200, 'saloon'] = 4
```
