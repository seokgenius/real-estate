```python
import pandas as pd
```

```python
waste_df = pd.read_csv('waste.csv', encoding='CP949')
waste_df.head()
```

```python
gu_list = []

for i in range(0, len(waste_df)):
    if type(waste_df['oldAddress'][i]) == float:
        addr = waste_df['newAddress'][i].split()
        gu_list.append(addr[1])
    else:
        addr = waste_df['oldAddress'][i].split()
        gu_list.append(addr[1])

print(gu_list)
```

```python
waste_df['waste'] = pd.Series(gu_list)
```

```python
waste_df
```

```python
waste_df['waste'].value_counts()
```

```python
waste_number = pd.DataFrame(waste_df['waste'].value_counts())
waste_number['region'] = waste_number.index
waste_number = waste_number[['region','waste']]
waste_number
```

```python
waste_number.to_csv('waste_raw.csv', index=False, encoding='utf-8-sig')
```

스케일 조정

- ~ 2 : 0
- 3 ~ 9 : 0.8
- 10 ~ 12 : 1.6
- 13 ~ 21 : 2.4
- 22 ~ 37 : 3.2
- 38 ~ : 4.0

```python
waste_number.loc[waste_number['waste'] <= 2, 'waste'] = 0
waste_number.loc[(waste_number['waste'] > 2) & (waste_number['waste'] <= 9) , 'waste'] = 0.8
waste_number.loc[(waste_number['waste'] > 9) & (waste_number['waste'] <= 12) , 'waste'] = 1.6
waste_number.loc[(waste_number['waste'] > 12) & (waste_number['waste'] <= 21) , 'waste'] = 2.4
waste_number.loc[(waste_number['waste'] > 22) & (waste_number['waste'] <= 37) , 'waste'] = 3.2
waste_number.loc[waste_number['waste'] > 37, 'waste'] = 4
```
