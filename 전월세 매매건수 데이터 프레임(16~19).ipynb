```python
rent16 = pd.read_csv('apart_rent16.csv', encoding='ANSI')
rent17 = pd.read_csv('apart_rent17.csv', encoding='ANSI')
rent18 = pd.read_csv('apart_rent18.csv', encoding='ANSI')
rent19 = pd.read_csv('apart_rent19.csv', encoding='ANSI')
```

```python
rent16.head()
```

```python
# 16년도 전월세 DataFrame
rent16_list = pd.DataFrame({
                '구' : rent16['시군구'],
                '전월세구분' : rent16['전월세구분'],
                '계약년월' : rent16['계약년월'],
                '전용면적(㎡)' : rent16['전용면적(㎡)'],                 
                '거래금액(만원)' : rent16['계약일'],
                '보증금(만원)' : rent16['보증금(만원)']
})

rent16_list['구'] = [each_gu.split()[1] for each_gu in rent16['시군구']]

# object -> list
rent16_list['보증금(만원)'] = rent16_list['보증금(만원)'].replace('[\,]', '', regex=True)
rent16_list['보증금(만원)'] = pd.to_numeric(rent16_list['보증금(만원)'])


rent16_list.info()


# 17년도 전월세 DataFrame
rent17_list = pd.DataFrame({
                '구' : rent17['시군구'],
                '전월세구분' : rent17['전월세구분'],
                '계약년월' : rent17['계약년월'],
                '전용면적(㎡)' : rent17['전용면적(㎡)'],                 
                '거래금액(만원)' : rent17['계약일'],
                '보증금(만원)' : rent17['보증금(만원)']
})

rent17_list['구'] = [each_gu.split()[1] for each_gu in rent17['시군구']]

# object -> list
rent17_list['보증금(만원)'] = rent17_list['보증금(만원)'].replace('[\,]', '', regex=True)
rent17_list['보증금(만원)'] = pd.to_numeric(rent17_list['보증금(만원)'])


# 18년도 전월세 DataFrame
rent18_list = pd.DataFrame({
                '구' : rent18['시군구'],
                '전월세구분' : rent18['전월세구분'],
                '계약년월' : rent18['계약년월'],
                '전용면적(㎡)' : rent18['전용면적(㎡)'],                 
                '거래금액(만원)' : rent18['계약일'],
                '보증금(만원)' : rent18['보증금(만원)']
})

rent18_list['구'] = [each_gu.split()[1] for each_gu in rent18['시군구']]

# object -> list
rent18_list['보증금(만원)'] = rent18_list['보증금(만원)'].replace('[\,]', '', regex=True)
rent18_list['보증금(만원)'] = pd.to_numeric(rent18_list['보증금(만원)'])




# 19년도 전월세 DataFrame
rent19_list = pd.DataFrame({
                '구' : rent19['시군구'],
                '전월세구분' : rent19['전월세구분'],
                '계약년월' : rent19['계약년월'],
                '전용면적(㎡)' : rent19['전용면적(㎡)'],                 
                '거래금액(만원)' : rent19['계약일'],
                '보증금(만원)' : rent19['보증금(만원)']
})

rent19_list['구'] = [each_gu.split()[1] for each_gu in rent19['시군구']]

# object -> list
rent19_list['보증금(만원)'] = rent19_list['보증금(만원)'].replace('[\,]', '', regex=True)
rent19_list['보증금(만원)'] = pd.to_numeric(rent19_list['보증금(만원)'])
```

```python
# 전세면 0, 월세면 1
rent_map = { 
    '전세' : 0,
    '월세' : 1

}
rent16_list['전월세구분'] = rent16_list['전월세구분'].map(rent_map)
rent17_list['전월세구분'] = rent17_list['전월세구분'].map(rent_map)
rent18_list['전월세구분'] = rent18_list['전월세구분'].map(rent_map)
rent19_list['전월세구분'] = rent19_list['전월세구분'].map(rent_map)
```

```python
# 구별 데이터 그룹화

# 16년도 데이터
rent16_list= pd.DataFrame({
                '구' : rent16_list['구'],
                '전월세구분' : rent16_list['전월세구분']              
})

rent16_list= rent16_list.groupby(['구','전월세구분']).agg({'전월세구분' : 'count'})
rent16_list.rename(columns = {'전월세구분' : '전월세건수'}, inplace = True) 



# 17년도 데이터
rent17_list= pd.DataFrame({
                '구' : rent17_list['구'],
                '전월세구분' : rent17_list['전월세구분']              
})

rent17_list= rent17_list.groupby(['구', '전월세구분']).agg({'전월세구분' : 'count'})
rent17_list.rename(columns = {'전월세구분' : '전월세건수'}, inplace = True) 


# 18년도 데이터
rent18_list= pd.DataFrame({
                '구' : rent18_list['구'],
                '전월세구분' : rent18_list['전월세구분']              
})

rent18_list= rent18_list.groupby(['구','전월세구분']).agg({'전월세구분' : 'count'})
rent18_list.rename(columns = {'전월세구분' : '전월세건수'}, inplace = True) 


# 19년도 데이터
rent19_list= pd.DataFrame({
                '구' : rent19_list['구'],
                '전월세구분' : rent19_list['전월세구분']              
})

rent19_list= rent19_list.groupby(['구', '전월세구분']).agg({'전월세구분' : 'count'})
rent19_list.rename(columns = {'전월세구분' : '전월세건수'}, inplace = True) 

```
