```python
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib import pyplot as plt

from matplotlib import rcParams
plt.rc('font', family='Malgun Gothic')
plt.rc('font', size='10')
```

```python
rent16_count = pd.read_csv('rent16_count.csv', encoding='ANSI')
rent17_count = pd.read_csv('rent17_count.csv', encoding='ANSI')
rent18_count = pd.read_csv('rent18_count.csv', encoding='ANSI')
rent19_count = pd.read_csv('rent19_count.csv', encoding='ANSI')
```

```python
# 구별 월세 건수를 출력
# 16년도 월세 정보
rent16_count['전월세구분']
rent16_count1_1 = rent16_count['전월세구분'] == 1
rent16_count2_1 = rent16_count[rent16_count1_1]

# 17년도 월세 정보
rent17_count['전월세구분']
rent17_count1_1 = rent17_count['전월세구분'] == 1
rent17_count2_1 = rent17_count[rent17_count1_1]

# 18년도 월세 정보
rent18_count['전월세구분']
rent18_count1_1 = rent18_count['전월세구분'] == 1
rent18_count2_1 = rent18_count[rent18_count1_1]

# 19년도 월세 정보
rent19_count['전월세구분']
rent19_count1_1 = rent19_count['전월세구분'] == 1
rent19_count2_1 = rent19_count[rent19_count1_1]
rent16_count2_1

```

```python
# 구 이름을 숫자로 대응시킨 딕셔너리
# 인덱싱을 하기 위해 필요
gu_dict = {
    '강남구':0,
    '강동구':1,
    '강북구':2,
    '강서구':3,
    '관악구':4,
    '광진구':5,
    '구로구':6,
    '금천구':7,
    '노원구':8,
    '도봉구':9,
    '동대문구':10,
    '동작구':11,
    '마포구':12,
    '서대문구':13,
    '서초구':14,
    '성동구':15,
    '성북구':16,
    '송파구':17,
    '양천구':18,
    '영등포구':19,
    '용산구':20,
    '은평구':21,
    '종로구':22,
    '중구':23,
    '중랑구':24
}
gu_dict
```

```python
# 구와 전월세 선택시 년도별 계약건수 출력. 
def gu_graph() :
    gu = input('구를 입력하세요 : ')
    rent = input('0.전세 1.월세: ')
    if rent == '0' :
        fig = plt.figure(figsize = (10, 4))
        ax2 = fig.add_subplot(111)
        cont_list = [rent16_count2.전월세건수[gu_dict[gu]],rent17_count2.전월세건수[gu_dict[gu]],rent18_count2.전월세건수[gu_dict[gu]],rent19_count2.전월세건수[gu_dict[gu]]]
        ax2.plot(['2016','2017','2018','2019'], cont_list)
        ax2.set_title(f"{rent16_count2['구'][gu_dict[gu]]} 전세계약")
        ax2.set_ylabel('매매건수')
        plt.show()
                      
    elif rent == '1' : 
        fig = plt.figure(figsize = (10, 4))
        ax2 = fig.add_subplot(111)
        cont_list = [rent16_count2_1.전월세건수[gu_dict[gu]],rent17_count2_1.전월세건수[gu_dict[gu]],rent18_count2_1.전월세건수[gu_dict[gu]],rent19_count2_1.전월세건수[gu_dict[gu]]]
        ax2.plot(['2016','2017','2018','2019'], cont_list)
        ax2.set_title(f"{rent16_count2_1['구'][gu_dict[gu]]} 월세계약")
        ax2.set_ylabel('매매건수')
        plt.show()
    else : 
        print('Error')
```

```python
gu_graph()
```

```python
# 16 ~19년도 월세 건수. 
fig = plt.figure(figsize = (20,4))
ax = fig.add_subplot(111)

ax.plot(rent16_count2_1['구'], rent16_count2_1.전월세건수, label='16년도 월세 계약건수')
ax.plot(rent17_count2_1['구'], rent17_count2_1.전월세건수, label='17년도 월세 계약건수')
ax.plot(rent18_count2_1['구'], rent18_count2_1.전월세건수, label='18년도 월세 계약건수')
ax.plot(rent19_count2_1['구'], rent19_count2_1.전월세건수, label='19년도 월세 계약건수')

ax.set_title('전세거래량')
ax.set_ylabel('구')
ax.set_ylabel('매매건수')

ax.legend()
plt.show()
```
